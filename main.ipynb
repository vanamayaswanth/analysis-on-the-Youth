```python
import pandas as pd 
```

```python
data=pd.read_csv('adult.csv')
df=data.copy(deep=True)
df.head()
```

```python
#add a colum name to the dataframe
df.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','income']
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['workclass']=le.fit_transform(df['workclass'])
df['education']=le.fit_transform(df['education'])
df['marital-status']=le.fit_transform(df['marital-status'])
df['occupation']=le.fit_transform(df['occupation'])
df['relationship']=le.fit_transform(df['relationship'])
df['race']=le.fit_transform(df['race'])
df['sex']=le.fit_transform(df['sex'])
df['native_country']=le.fit_transform(df['native_country'])
df['income']=le.fit_transform(df['income'])
df.head()
```

```python
#train and test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('income',axis=1),df['income'],test_size=0.2,random_state=0)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)

#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
print(accuracy_score(y_test,y_pred))

#f1 score
from sklearn.metrics import f1_score
print(f1_score(y_test,y_pred))

#recall
from sklearn.metrics import recall_score
print(recall_score(y_test,y_pred))

#precision
from sklearn.metrics import precision_score
print(precision_score(y_test,y_pred))

#confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))

#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred_rf=rf.predict(X_test)

#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_rf)
print(accuracy_score(y_test,y_pred_rf))

#f1 score
from sklearn.metrics import f1_score
print(f1_score(y_test,y_pred_rf))

#recall
from sklearn.metrics import recall_score
print(recall_score(y_test,y_pred_rf))

#precision
from sklearn.metrics import precision_score
print(precision_score(y_test,y_pred_rf))

#confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred_rf))

#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_rf))
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
y_pred_lr=lr.predict(X_test)

#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_lr)
print(accuracy_score(y_test,y_pred_lr))

#f1 score
from sklearn.metrics import f1_score
print(f1_score(y_test,y_pred_lr))

#recall
from sklearn.metrics import recall_score
print(recall_score(y_test,y_pred_lr))

#precision
from sklearn.metrics import precision_score
print(precision_score(y_test,y_pred_lr))


#confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred_lr))

#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_lr))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred_knn=knn.predict(X_test)

#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_knn)
print(accuracy_score(y_test,y_pred_knn))


#f1 score
from sklearn.metrics import f1_score
print(f1_score(y_test,y_pred_knn))


#recall
from sklearn.metrics import recall_score
print(recall_score(y_test,y_pred_knn))


#precision
from sklearn.metrics import precision_score
print(precision_score(y_test,y_pred_knn))


#confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred_knn))


#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_knn))
```

```python
#svm
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_train,y_train)
y_pred_svm=svm.predict(X_test)

#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_svm)

#f1 score
from sklearn.metrics import f1_score
print(f1_score(y_test,y_pred_svm))

#recall
from sklearn.metrics import recall_score
print(recall_score(y_test,y_pred_svm))

#precision
from sklearn.metrics import precision_score
print(precision_score(y_test,y_pred_svm))

#confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred_svm))

#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_svm))
```
